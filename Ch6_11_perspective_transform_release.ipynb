- `Affine Transform: All target points exists on the same plane with` $z=1 (w=1)$  
- `Perspective Transform: All target points can be moved on the different planes with different` $z (w) $ `values`


### How to calculate Perspective transform matrix

$$
\begin{aligned}
\left[
\begin{matrix}
    w_1x_1^{\prime}\\
    w_1y_1^{\prime}\\
    w_1
\end{matrix}
\right] =
\left[
\begin{matrix}
    a & b & c \\
    d & e & f \\
    g & h & 1 \\    
\end{matrix}
\right]
\left[
    \begin{matrix}
    x_1\\
    y_1\\
    1
\end{matrix}
\right]
\end{aligned}
$$

- It gives

$$w_1x_1^{\prime} = ax_1 + by_1 + c$$

$$w_1y_1^{\prime} = dx_1 + ey_1 + f$$

$$w_1 = gx_1 + hy_1 + 1$$

- $x_1^{\prime}$ and $y_1^{\prime}$ can be obtained by dividing by $w_1$

$$x_1^{\prime} = \frac{w_1x_1^{\prime}}{w_1} = \frac{ax_1 + by_1 + c}{gx_1 + hy_1 + 1}\rightarrow x_1^{\prime}(gx_1 + hy_1 + 1) = ax_1 + by_1 + c$$

$$y_1^{\prime} = \frac{w_1y_1^{\prime}}{w_1} = \frac{dx_1 + ey_1 + f}{gx_1 + hy_1 + 1}\rightarrow y_1^{\prime}(gx_1 + hy_1 + 1) = dx_1 + ey_1 + f$$

- which can be rewritten as

$$x_1^{\prime} = ax_1 + by_1 + c - gx_1x_1^{\prime} - hx_1^{\prime}y_1$$

$$y_1^{\prime} = dx_1 + ey_1 + f - gx_1y_1^{\prime} - hy_1y_1^{\prime}$$



### 8 unknown variables $(a, b, c, \cdots, h)$
- `4 points are thus needed.`


$
\begin{aligned}
\left[
    \begin{matrix}
    x_1^{\prime}\\
    y_1^{\prime}\\
    x_2^{\prime}\\
    y_2^{\prime}\\
    x_3^{\prime}\\
    y_3^{\prime}\\
    x_4^{\prime}\\
    y_4^{\prime}\\
\end{matrix}
\right] =
\left[
    \begin{matrix}
    x_1 & y_1 &1 &0 & 0 &0 &-x_1x_1^{\prime}&-x_1^{\prime}y_1\\
    0 & 0 & 0 &x_1 & y_1 & 1 &-x_1y_1^{\prime}&-y_1y_1^{\prime}\\
    x_2 & y_2 &1 &0 & 0 &0 &-x_2x_2^{\prime}&-x_2^{\prime}y_2\\
    0 & 0 &0 &x_2 & y_2 &1 &-x_2y_2^{\prime}&-y_2y_2^{\prime}\\
    x_3 & y_3 &1 &0 & 0 &0 &-x_3x_3^{\prime}&-x_3^{\prime}y_3\\
    0 & 0 &0 &x_3 & y_3 &1 &-x_3y_3^{\prime}&-y_3y_3^{\prime}\\
    x_4 & y_4 &1 &0 & 0 &0 &-x_4x_4^{\prime}&-x_4^{\prime}y_4\\
    0 & 0 &0 &x_4 & y_4 &1 &-x_4y_4^{\prime}&-y_4y_4^{\prime}
\end{matrix}
\right]
\left[
    \begin{matrix}
    a\\
    b\\
    c\\
    d\\
    e\\
    f\\
    g\\
    h\\
\end{matrix}
\right]
\end{aligned}
$

- The perspective matrix elements can be obtained by matrix inversion

$
\begin{aligned}
\left[
    \begin{matrix}
    a\\
    b\\
    c\\
    d\\
    e\\
    f\\
    g\\
    h\\
\end{matrix}
\right]=
\left[
    \begin{matrix}
    x_1 & y_1 &1 &0 & 0 &0 &-x_1x_1^{\prime}&-x_1^{\prime}y_1\\
    0 & 0 & 0 &x_1 & y_1 & 1 &-x_1y_1^{\prime}&-y_1y_1^{\prime}\\
    x_2 & y_2 &1 &0 & 0 &0 &-x_2x_2^{\prime}&-x_2^{\prime}y_2\\
    0 & 0 &0 &x_2 & y_2 &1 &-x_2y_2^{\prime}&-y_2y_2^{\prime}\\
    x_3 & y_3 &1 &0 & 0 &0 &-x_3x_3^{\prime}&-x_3^{\prime}y_3\\
    0 & 0 &0 &x_3 & y_3 &1 &-x_3y_3^{\prime}&-y_3y_3^{\prime}\\
    x_4 & y_4 &1 &0 & 0 &0 &-x_4x_4^{\prime}&-x_4^{\prime}y_4\\
    0 & 0 &0 &x_4 & y_4 &1 &-x_4y_4^{\prime}&-y_4y_4^{\prime}
\end{matrix}
\right]^{-1}
\left[
    \begin{matrix}
    x_1^{\prime}\\
    y_1^{\prime}\\
    x_2^{\prime}\\
    y_2^{\prime}\\
    x_3^{\prime}\\
    y_3^{\prime}\\
    x_4^{\prime}\\
    y_4^{\prime}\\
\end{matrix}
\right]
\end{aligned}
$



### An example for specific points

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
src = cv2.imread('../images/railroad.jpg')
src_copy = src.copy()
H, W = src.shape[:2]

#L.TOP->L.BOT->R.TOP->R.BOT
pts_src = np.array( [[297,0], [13, H-1], [500,0],  [766, H-1] ] )
pts_dst = np.array( [[0,0], [0, H-1], [W-1, 0], [W-1, H-1]] )

#*************Perspective matrix calculation
x1 = pts_src[0,0]
y1 = pts_src[0,1]
x2 = pts_src[1,0]
y2 = pts_src[1,1]
x3 = pts_src[2,0]
y3 = pts_src[2,1]
x4 = pts_src[3,0]
y4 = pts_src[3,1]

u1 = pts_dst[0,0]
v1 = pts_dst[0,1]
u2 = pts_dst[1,0]
v2 = pts_dst[1,1]
u3 = pts_dst[2,0]
v3 = pts_dst[2,1]
u4 = pts_dst[3,0]
v4 = pts_dst[3,1]

a = [
    [x1, y1, 1,  0,  0,  0,  -x1*u1, -y1*u1],
    [0,  0,  0,  x1, y1, 1,  -x1*v1, -y1*v1],
    [x2, y2, 1,  0,  0,  0,  -x2*u2, -y2*u2],
    [0,  0,  0,  x2, y2, 1,  -x2*v2, -y2*v2],
    [x3, y3, 1,  0,  0,  0,  -x3*u3, -y3*u3],
    [0,  0,  0,  x3, y3, 1,  -x3*v3, -y3*v3],
    [x4, y4, 1,  0,  0,  0,  -x4*u4, -y4*u4],
    [0,  0,  0,  x4, y4, 1,  -x4*v4, -y4*v4],
    ]

a = np.array(a)
u = np.array( [ [u1,v1,u2,v2,u3,v3,u4,v4] ]).T

m = np.matmul(np.linalg.inv(a), u)
m = m[:,0]
mat = [
    [m[0], m[1], m[2]],
    [m[3], m[4], m[5]],
    [m[6], m[7], 1],
    ]

print('\n pts_src')
print(np.array(pts_src).T )

print('\n pts_dst')
print(np.array(pts_dst).T)


mat = np.array(mat)
print(f'\n Matrix_Calculation\n {mat}')

mat_cv2 = cv2.getPerspectiveTransform(np.float32( pts_src ), np.float32( pts_dst) )
print(f'\n Matrix_CV2\n {mat_cv2}')


pts_src_3d = np.array( [ [x1, y1, 1], [x2, y2, 1], [x3, y3, 1], [x4, y4, 1]] ).T
pts_dst_3d = np.matmul(mat_cv2, pts_src_3d)

print('\n pts_dst_3d')
print(pts_dst_3d)


print('\n Scaling')
print(pts_dst_3d[2])

print('\n pts_dst_3d_original')
print(pts_dst_3d/pts_dst_3d[2])


#********************************

for pt in pts_src:
    cv2.circle(src_copy, pt, 10, (0,0,255), -1)

#*********** points should be float32!!!
matrix_transform = cv2.getPerspectiveTransform(pts_src.astype(np.float32), pts_dst.astype(np.float32))
dsize = ( W, H)
dst = cv2.warpPerspective(src, matrix_transform, dsize)
for pt in pts_dst:
    cv2.circle(dst, pt, 10, (0,0,255), -1)
# cv2.imshow('SRC', src_copy)
# cv2.imshow('DST', dst)

final = np.concatenate((src_copy, dst), axis=1)
plt.figure(figsize=(20,10))
plt.imshow(final[:,:,::-1])
plt.show()
```
