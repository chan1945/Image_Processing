## 1. How to change coordinates between X-Y and Array

- `X-Y coordinate and image matrix use different indices`

- **`(h,w)`** $\color{yellow}\boldsymbol{\rightarrow}$ **`(x,y)`**
    - 1. Switch $h$ and $w$
    - 2. Change the sign of $y$
    - 3. Move to the center of images to (0,0)
        - $x \gets x-\frac{W}{2} $
        - $y \gets y-\frac{H}{2} $

- $ [H, W] \rightarrow \left(\frac{W}{2},-\frac{H}{2}\right)$

무조건 시험에 나옴

```python
H, W = 100, 200

def ij2xy(pt):
    # 1. switch coordinates
    y,x = pt 


    # 2. change of sign y
    y = -y

    # 3. move
    x = x-W/2
    y = y-H/2

    return (int(x), int(y))

def xy2ij(pt):
    x,y = pt 

    # 1. move
    x = x + W/2
    y = y + H/2

    # 2. change of sign y
    y = -y
    
    # 3. switch coordinates
    h = int(y)
    w = int(x)
    return (h,w)
```

```python
list(map(xy2ij, list( map(ij2xy, ((0,0), (0,W), (H,0), (H,W)) ) ) ) )
```
