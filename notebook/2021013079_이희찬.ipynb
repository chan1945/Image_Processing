```python
# Q3 - 1
A = np.full((200,200), 255, dtype = np.uint8)
A[0:100, :] = 0

B = np.full((200,200), 255, dtype = np.uint8)
B[:, 100:200] = 0

plt.subplot(1,2,1)
plt.imshow(A , cmap ='gray')
plt.subplot(1,2,2)
plt.imshow(B , cmap = 'gray')
plt.show()
```

```python
# Q3 - 2
dst1 = A & B

dst2 = A | B

dst3 = ~A ^ ~B 

plt.subplot(1,3,1)
plt.imshow(dst1, cmap = 'gray')
plt.subplot(1,3,2)
plt.imshow(dst2 , cmap = 'gray')
plt.subplot(1,3,3)
plt.imshow(dst3, cmap = 'gray')
plt.show()
```

```python
# Q4 - 1
H = 70
W = 50

src = np.zeros((H, W), dtype = np.uint8)
for i in range(src.shape[0]):
    for j in range(src.shape[1]):
        src[i, j] = min(i * 5 +j , 255)
        
plt.imshow(src, cmap = 'gray')
plt.show()

hist, bins = np.histogram(src, bins = 256, range = (0, 256))
hist_cum = np.cumsum(hist)

plt.subplot(1,2,1)
plt.bar(bins[1:], hist)
plt.subplot(1,2,2)
plt.bar(bins[1:], hist_cum)
plt.show()
```

```python
# Q4 -2

# lut 생성
size = src.shape[0] * src.shape[1]
eq_lut = hist_cum * (1/size) * 255
# 맵핑
src_eq = eq_lut[src]

plt.subplot(1,2,1)
plt.imshow(src, cmap = 'gray')
plt.subplot(1,2,2)
plt.imshow(src_eq, cmap = 'gray')
plt.show()
```
