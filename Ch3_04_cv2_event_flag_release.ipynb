# Mouse Callback in CV2


해당 내용은 시험에 안나옴

- Types of `events`
    - `Impossible to have multiple events simultaneoulsy`

<nop/> <div style="width:800px">

- Types of `flags`
    - `마우스 이벤트가 발생할 때 키보드 또는 마우스 현재 상태`. Shif+마우스 등
    - `Bitmap`: Single bit is only set to 1
    - `bitwise AND (&)` is used to check multiple flags simultaneously:
        - `(flags & cv2.EVENT_FLAG_CTRLKEY) and (flags & cv2.EVENT_FLAG_SHIFTKEY)`



```python
import numpy as np
import cv2

#  event, x, y, flags, param: 순서대로 입력
# param:전달하고 싶은 데이터.
def mouse_callback(event, x, y, flags, param):
    print(f'event={event}, flags={flags}')
    if event == cv2.EVENT_LBUTTONDOWN:
        print(f'Left Down:({x},{y})')
    elif event == cv2.EVENT_LBUTTONUP:
        print(f'Left Up:({x},{y})')
    elif event == cv2.EVENT_RBUTTONDOWN:
        print(f'Right Down:({x},{y})')
    elif event == cv2.EVENT_RBUTTONUP:
        print(f'Right Up:({x},{y})')
    elif event == cv2.EVENT_LBUTTONDBLCLK:
        print(f'Left Double Click:({x},{y})')
    elif event == cv2.EVENT_RBUTTONDBLCLK:
        print(f'Right Double Click:({x},{y})')
    elif event == cv2.EVENT_MOUSEMOVE: # 마우스가 움직일 때 발생
        print(f'Mouse is moving:({x},{y})')
        

img = np.full((600, 800, 3), 255, dtype=np.uint8)
cv2.namedWindow('image')
# 마우스 입력, namedWIndow or imshow가 실행되어 창이 떠있는 상태에서만 사용가능
# 마우스 이벤트가 발생하면 mouse_callback 함수 실행
cv2.setMouseCallback('image', mouse_callback, img)

cv2.imshow('image', img)

cv2.waitKey()

cv2.destroyAllWindows()
```

### Drawing with mouse
- Move with the left button clicked to draw
    
- `Right Click` will remove all

> How to draw a line using OPEN-CV

```python
import numpy as np
import cv2


img = np.full((600, 800, 3), 255, dtype=np.uint8)
cv2.namedWindow('image')

cv2.imshow('image', img)
cv2.waitKey(1000)

cv2.line(img, (0, 0), (400, 300), (0, 0, 255), 4, cv2.LINE_AA)
cv2.imshow('image', img)
cv2.waitKey(1000)

cv2.line(img, (400, 300), (800, 0), (0, 0, 255), 4, cv2.LINE_AA)
cv2.imshow('image', img)
cv2.waitKey(1000)

cv2.line(img, (400, 300), (0, 600), (0, 0, 255), 4, cv2.LINE_AA)
cv2.imshow('image', img)
cv2.waitKey(1000)

cv2.line(img, (400, 300), (800, 600), (0, 0, 255), 4, cv2.LINE_AA)
cv2.imshow('image', img)
cv2.waitKey(1000)




cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

oldx = oldy = 0

def on_mouse(event, x, y, flags, param):
    global oldx, oldy, img

    if event == cv2.EVENT_MOUSEMOVE:
        if (event == cv2.EVENT_MOUSEMOVE) and (flags == cv2.EVENT_FLAG_LBUTTON):
            cv2.line(img, (oldx, oldy), (x, y), (0, 0, 255), 4, cv2.LINE_AA)
            cv2.imshow('image', img)
        if flags == cv2.EVENT_FLAG_RBUTTON:
            img[y -10:y+10, x-10:x+10] = 255
            cv2.imshow('image', img)
    

    # if event == cv2.EVENT_RBUTTONDOWN:
    #     img = np.ones((600, 800, 3), dtype=np.uint8) * 255
    #     cv2.imshow('image', img)
    oldx, oldy = x, y

img = np.ones((600, 800, 3), dtype=np.uint8) * 255
cv2.namedWindow('image')
cv2.setMouseCallback('image', on_mouse)
cv2.imshow('image', img)
cv2.waitKey()
cv2.destroyAllWindows()

```
